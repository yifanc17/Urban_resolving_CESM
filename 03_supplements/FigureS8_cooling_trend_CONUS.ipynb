```python
conus_ctrl = clip_netcdf_by_shp(ctrl_JJA_avg_1d32deg['TSKIN_U'],us_bounds)
conus_test = clip_netcdf_by_shp(test_JJA_avg_1d32deg['TSKIN_U'],us_bounds)

valid_mask = np.isfinite(conus_ctrl) & np.isfinite(conus_test)
conus_ctrl = conus_ctrl.where(valid_mask)
conus_test = conus_test.where(valid_mask)

datasets = [conus_ctrl, conus_test, conus_test-conus_ctrl]
dataset_names = ["CTRL_UHR", "USURF_UHR", "USURF-CTRL"]

plot_config = {
    'vmin': 290,         
    'vmax': 310,
    'cmap': 'RdYlBu_r',
    'figsize': (7, 12),
    'conus_extent': [-125, -66.5, 24, 50]
}

fig, axs = plt.subplots(3, 1, 
                       figsize=plot_config['figsize'], 
                       subplot_kw={'projection': ccrs.PlateCarree(central_longitude=0)})

for ax, data, title in zip(axs, datasets, dataset_names):
    for spine in ax.spines.values():
        spine.set_visible(False)
    if title == 'USURF-CTRL':
        vmin = -4
        vmax = 4
        fg = data.plot(
            ax=ax,
            transform=ccrs.PlateCarree(),
            cmap='RdYlBu_r', 
            vmax=vmax,
            vmin=vmin,
            alpha=1,
            add_colorbar=True,
            cbar_kwargs={'extend': 'both', 'pad': 0.05, 'shrink': 0.8, 'aspect': 20}
        )
        cbar = fg.colorbar
        cbar.ax.tick_params(labelsize=14)
        cbar.set_label("ΔTSKIN_U", fontsize=16)
    else:
        vmin = 290
        vmax = 310
        fg = data.plot(
            ax=ax,
            transform=ccrs.PlateCarree(),
            cmap='RdYlBu_r', 
            vmax=vmax,
            vmin=vmin,
            alpha=1,
            add_colorbar=True,
            cbar_kwargs={'extend': 'both', 'pad': 0.05, 'shrink': 0.8, 'aspect': 20}
        )
        cbar = fg.colorbar
        cbar.ax.tick_params(labelsize=14)
        cbar.set_label("TSKIN_U", fontsize=16)
    ax.coastlines(color='black', linewidth=0.8)
    ax.add_feature(cfeature.BORDERS, edgecolor='gray', linewidth=0.5, zorder=150)
    
    gl = ax.gridlines(draw_labels=False, color='white', linewidth=1.2, linestyle='--', zorder=3)
    gl.xlocator = plt.FixedLocator([-100, 0, 100])
    gl.ylocator = plt.FixedLocator([-90, -45, 0, 45, 90])
    
    ax.add_feature(cfeature.OCEAN, color='whitesmoke', zorder=1)
    ax.add_feature(cfeature.LAKES, edgecolor='gray', facecolor='whitesmoke', linewidth=0.5, zorder=50)

    ax.set_title(title, fontsize=20)
    ax.set_aspect('auto')


fig.tight_layout() 
fig.savefig(plot_path+'TSKINU_2010_2014_JJA_CTRL_TEST_DIFF_CONUS_c250711.png', format='png', bbox_inches='tight', dpi=600)

for data in datasets:
    print(np.nanmean(data), np.nanmin(data), np.nanmax(data))
```
