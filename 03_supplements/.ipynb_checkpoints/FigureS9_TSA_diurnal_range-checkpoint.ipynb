```python
def prepare_data(data_array):
    array = data_array.values.flatten()
    array = array[~np.isnan(array)]
    array = array[array>0]
    return array

def compute_boxplot_stats(data):
    stats = {
        'med': np.percentile(data, 50),
        'q1': np.percentile(data, 25),
        'q3': np.percentile(data, 75),
        'whislo': np.percentile(data, 1),   
        'whishi': np.percentile(data, 99),
        'mean': np.nanmean(data),
        'std': np.nanstd(data)
    }
    return stats
```

```python
def extract_diurnal_range_10cities(period='JJA'): # remove PRISM from plots
    ctrl_Tmin=(eval(f'ctrl_{period}_avg_1d32deg_reg')['TREFMNAV_U'])
    test_Tmin=(eval(f'test_{period}_avg_1d32deg_reg')['TREFMNAV_U'])
    obs_Tmin=PRISM_JJA_Tmin_xr+273.15
    mask=(ctrl_Tmin>0) & (obs_Tmin>0) & (test_Tmin>0)
    ctrl_Tmin=ctrl_Tmin.where(mask)
    test_Tmin=test_Tmin.where(mask)
    obs_Tmin=obs_Tmin.where(mask)

    ctrl_Tmax=(eval(f'ctrl_{period}_avg_1d32deg_reg')['TREFMXAV_U'])
    test_Tmax=(eval(f'test_{period}_avg_1d32deg_reg')['TREFMXAV_U'])
    obs_Tmax=PRISM_JJA_Tmax_xr+273.15
    mask=(ctrl_Tmax>0) & (obs_Tmax>0) & (test_Tmax>0)
    ctrl_Tmax=ctrl_Tmax.where(mask)
    test_Tmax=test_Tmax.where(mask)
    obs_Tmax=obs_Tmax.where(mask)
    
    city_array = {}
    city_stats = {}
    
    datasets = [
        {'name': 'ctrl_Tmin', 'variable': ctrl_Tmin, 'color': '#9fc5e8'},
        {'name': 'test_Tmin', 'variable': test_Tmin, 'color': '#f4cccc'},
        {'name': 'obs_Tmin', 'variable': obs_Tmin, 'color': '#93c47d'},
        {'name': 'ctrl_Tmax', 'variable': ctrl_Tmax, 'color': '#9fc5e8'},
        {'name': 'test_Tmax', 'variable': test_Tmax, 'color': '#f4cccc'},
        {'name': 'obs_Tmax', 'variable': obs_Tmax, 'color': '#93c47d'}
    ]
    
    all_boxplot_data = []
    colors_for_boxes = ['#9fc5e8','#f4cccc']*10
    positions = []
    city_positions = []
    pos = 1 
    
    for city in PRISM_cities:
        city_name = city['name']
        city_shapefile = city['shapefile']
        city_waterFrac = city['water']
        for dataset in datasets:
            dataset_name = dataset['name']
            dataset_variable = dataset['variable']
            color = dataset['color']
    
            clipped_data = clip_netcdf_by_shp(dataset_variable, city_shapefile).where(city_waterFrac>0.0)
            data_array = prepare_data(clipped_data)
            city_array[(city_name,dataset_name)]=data_array
                    
        ctrl_Tmin=city_array.get((city_name, 'ctrl_Tmin'))
        test_Tmin=city_array.get((city_name, 'test_Tmin'))
        obs_Tmin=city_array.get((city_name, 'obs_Tmin'))
        mask=(~np.isnan(ctrl_Tmin)) & (~np.isnan(obs_Tmin)) & (~np.isnan(test_Tmin))
        ctrl_Tmin = ctrl_Tmin[mask].flatten()
        test_Tmin = test_Tmin[mask].flatten()
        obs_Tmin = obs_Tmin[mask].flatten()
        
        ctrl_Tmax=city_array.get((city_name, 'ctrl_Tmax'))
        test_Tmax=city_array.get((city_name, 'test_Tmax'))
        obs_Tmax=city_array.get((city_name, 'obs_Tmax'))
        mask=(~np.isnan(ctrl_Tmax)) & (~np.isnan(obs_Tmax)) & (~np.isnan(test_Tmax))
        ctrl_Tmax = ctrl_Tmax[mask].flatten()
        test_Tmax = test_Tmax[mask].flatten()
        obs_Tmax = obs_Tmax[mask].flatten()

        ctrl_range=ctrl_Tmax-ctrl_Tmin
        test_range=test_Tmax-test_Tmin
        obs_range=obs_Tmax-obs_Tmin
            
        MAE_ctrl_range = mean_absolute_error(obs_range, ctrl_range)
        MAE_test_range = mean_absolute_error(obs_range, test_range)
        diff_range = np.mean(test_range-ctrl_range)
        
        city_stats[city_name] = {
                        'mean_ctrl_range': np.nanmean(ctrl_range),
                        'mean_test_range': np.nanmean(test_range),
                        'mean_diff_range': diff_range
                    }
        for data_array in [ctrl_range, test_range]:
            stats = compute_boxplot_stats(data_array)
            box_data = {
                'med': stats['med'],
                'q1': stats['q1'],
                'q3': stats['q3'],
                'whislo': stats['whislo'],
                'whishi': stats['whishi'],
                'fliers': []
            }
            all_boxplot_data.append(box_data)
            positions.append(pos)
            pos += 1

            
        city_positions.append(pos - 2)
        pos += 0.5

    return all_boxplot_data, colors_for_boxes, positions, city_positions, city_stats
```

```python
def boxplot_diurnal_range(period):
    all_boxplot_data, colors_for_boxes, positions, city_positions, city_stats=extract_diurnal_range_10cities(period)
    fig, ax = plt.subplots(figsize=(15, 8))
    boxplot = ax.bxp(all_boxplot_data, positions=positions, showfliers=False, patch_artist=True,widths=0.7)
    
    num_boxes = len(boxplot['boxes'])
    for i in range(num_boxes):
        box = boxplot['boxes'][i]
        whisker_left = boxplot['whiskers'][2 * i]
        whisker_right = boxplot['whiskers'][2 * i + 1]
        cap_left = boxplot['caps'][2 * i]
        cap_right = boxplot['caps'][2 * i + 1]
        median = boxplot['medians'][i]
        
        color = colors_for_boxes[i]
        
        box.set_facecolor('none')        
        box.set_edgecolor(color)         
        box.set_linewidth(3)          
        
        # Set whisker properties
        whisker_left.set_color(color)
        whisker_right.set_color(color)
        whisker_left.set_linewidth(3)
        whisker_right.set_linewidth(3)
        whisker_left.set_linestyle('dashed')    
        whisker_right.set_linestyle('dashed') 
        
        # Set cap properties
        cap_left.set_color(color)
        cap_right.set_color(color)
        cap_left.set_linewidth(3)
        cap_right.set_linewidth(3)
        
        median.set_color(color)
        median.set_linewidth(2.5)
    
    ax.set_xticks(city_positions)
    ax.set_xticklabels([city['name'] for city in PRISM_cities], ha='center',fontsize=28)
    
    ax.set_ylabel('Tmax-Tmin',fontsize=24)
    ax.tick_params(axis='y', labelsize=20)
    ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f'))
    ax.set_title('')
    
    legend_handles = [
        mpatches.Patch(facecolor='none', edgecolor='#9fc5e8', label='CTRL', linewidth=3),
        mpatches.Patch(facecolor='none', edgecolor='#f4cccc', label='USURF', linewidth=3)
    ]
    ax.legend(handles=legend_handles,fontsize=20)
    
    plt.tight_layout()

    city_stats = pd.DataFrame(city_stats)
    city_stats['average'] = city_stats.mean(axis=1)
    return fig,city_stats
```
