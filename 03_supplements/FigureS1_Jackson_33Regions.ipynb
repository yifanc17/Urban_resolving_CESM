```python
def convert_lsmlatlon_1deg(array): # align with CESM grids
    new_dims = {'lsmlat': 'lat', 'lsmlon': 'lon'}
    array = array.rename(new_dims)
    array.coords['lat'] = center_lats_1deg
    array = array.sortby(array.coords['lat'])
    array.coords['lon'] = center_lons_1deg
    array = array.sortby(array.coords['lon'])
    return array
```

```python
JK_surf = xr.open_dataset('/glade/campaign/cesm/cesmdata/inputdata/lnd/clm2/surfdata_map/surfdata_0.9x1.25_78pfts_simyr2000_c170428.nc')
urban_regions = JK_surf['URBAN_REGION_ID']
urban_regions = convert_lsmlatlon_1deg(urban_regions.where(urban_regions!=0)) # land pixels
colors = cm.nipy_spectral(np.linspace(0, 1, 33))
cmap = LinearSegmentedColormap.from_list('urban_regions', colors, N=33)
bounds = np.linspace(0, 33, 34)
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.subplot(projection=ccrs.PlateCarree(central_longitude=0.0))
cbar_kwargs = {
                  "orientation": "horizontal",
                  "shrink": 1,
                  "aspect":60,
                  "pad": 0.03,                   
                  "label": 'urban regions',
                  "spacing":'proportional',
                  "ticks": bounds[1:], 
                  "boundaries": bounds
              }

urban_regions.plot(ax=ax,
         transform=ccrs.PlateCarree(central_longitude=0),
         vmin = urban_regions.min(),
         vmax = urban_regions.max(),
         cmap=cmap,
         cbar_kwargs=cbar_kwargs)

ax.coastlines(linewidth=0.5, zorder=150)
ax.set_aspect('auto', adjustable=None)
ax.add_feature(cfeature.OCEAN, color='whitesmoke', zorder=1)
ax.add_feature(cfeature.LAKES, edgecolor='gray', facecolor='whitesmoke', linewidth=0.5, zorder=50)
ax.add_feature(cfeature.BORDERS, edgecolor='gray', linewidth=0.5, zorder=150)
```
