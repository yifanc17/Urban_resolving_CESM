```python
fig, axes = plt.subplots(3,1, figsize=(7,12), constrained_layout=True)

vmin = 0
vmax = 100
cmap = "viridis"

im0 = new_surf['PCT_URBAN'].sel(numurbl=2).where(lambda x: x > 0).plot(
    ax=axes[0], cmap=cmap, vmin=vmin, vmax=vmax, add_colorbar=False
)
cb0 = fig.colorbar(im0, ax=axes[0])
cb0.ax.tick_params(labelsize=14)
cb0.set_label("PCT_URBAN (%)", fontsize=16)

im1 = def_surf['PCT_URBAN'].sel(numurbl=2).where(lambda x: x > 0).plot(
    ax=axes[1], cmap=cmap, vmin=vmin, vmax=vmax, add_colorbar=False
)

cb1 = fig.colorbar(im1, ax=axes[1])
cb1.ax.tick_params(labelsize=14)
cb1.set_label("PCT_URBAN (%)", fontsize=16)

diff = new_surf['PCT_URBAN'].sel(numurbl=2) - def_surf['PCT_URBAN'].sel(numurbl=2)
im2 = diff.where(diff != 0).plot(
    ax=axes[2], cmap='coolwarm', vmin=-30, vmax=30, add_colorbar=False
)
cb2 = fig.colorbar(im2, ax=axes[2], extend='both')
cb2.ax.tick_params(labelsize=14)
cb2.set_label("ΔPCT_URBAN (%)", fontsize=16)

# Titles
axes[0].set_title(f"updated: #pixels={int((new_surf['PCT_URBAN'].sel(numurbl=2) > 0).sum())}", fontsize=18)
axes[1].set_title(f"default: #pixels={int((def_surf['PCT_URBAN'].sel(numurbl=2) > 0).sum())}", fontsize=18)
axes[2].set_title("updated - default", fontsize=18)

plt.show()
plt.tight_layout()
#fig.savefig(plot_path+"FigureS2_manually_adjusted_PCT_URBAN.png", dpi=400, bbox_inches='tight')
```
