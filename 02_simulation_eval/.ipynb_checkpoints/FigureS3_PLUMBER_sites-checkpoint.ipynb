```python
koppen_colors=['#f8dada','#fee3c6','#d9ead3','#b6d8f7','#f2edff']
koppen_cmap = mcolors.ListedColormap(koppen_colors)
koppen_bounds = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5]
koppen_norm = mcolors.BoundaryNorm(koppen_bounds, koppen_cmap.N)
```

```python
cmap = cm.get_cmap('gray_r') 
norm = mcolors.Normalize(vmin=0, vmax=1) 

climate_colors = {
    'Tropical': '#f8dada',
    'Dry': '#fee3c6',
    'Temperate': '#d9ead3',
    'Continental': '#b6d8f7',
    'Polar': '#f2edff'
}

def marker_size(period):
    if period > 1000:
        return 200
    elif 365 <= period <= 1000:
        return 100
    else:
        return 50

plumber_site['MarkerSize'] = plumber_site['Period'].apply(marker_size)
plumber_site['Color'] = plumber_site['Climate'].map(climate_colors)

fig = plt.figure(figsize=(16, 9))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()

ax.add_feature(cfeature.LAND, facecolor='#fff7e1')
ax.add_feature(cfeature.OCEAN, facecolor='white')
ax.coastlines()

koppen_simplified.where(koppen_simplified>0).plot(ax=ax,
                       transform=ccrs.PlateCarree(),
                        cmap=koppen_cmap, 
                        norm=koppen_norm,
                        shading='auto',
                        zorder=1,
                        alpha=1,
                        add_colorbar=False)

xticks = [-180, -90, 0, 90, 180] 
yticks = [-90, -45, 0, 45, 90]   

gl = ax.gridlines(
    draw_labels=True,
    linewidth=1,
    color='gray',
    alpha=0.5,
    linestyle=':'
)

gl.xlocator = plt.FixedLocator(xticks)
gl.ylocator = plt.FixedLocator(yticks)

gl.xlabel_style = {'size': 10}
gl.ylabel_style = {'size': 10}
gl.right_labels = False
gl.top_labels = False

for _, row in plumber_site.iterrows():
    lon = row['Longitude']
    lat = row['Latitude']
    color = row['Color']
    size = row['MarkerSize']
    frac = row['Impervious_fraction']

    inner_color = cmap(norm(frac))
    ax.plot(lon, lat, 'o', markersize=np.sqrt(size), color=inner_color,
            transform=ccrs.PlateCarree(), zorder=4, alpha=0.8)


climate_patches = [
    Patch(facecolor=color, edgecolor=None, label=climate)
    for climate, color in climate_colors.items()
]

size_legend = [
    Line2D([0], [0], marker='o', color='w', label='> 1000 days', markerfacecolor='gray', markersize=np.sqrt(200)),
    Line2D([0], [0], marker='o', color='w', label='365–1000 days', markerfacecolor='gray', markersize=np.sqrt(100)),
    Line2D([0], [0], marker='o', color='w', label='< 365 days', markerfacecolor='gray', markersize=np.sqrt(50))]

all_legend_elements = climate_patches + size_legend
legend = ax.legend(handles=all_legend_elements, loc='lower left', bbox_to_anchor=(0.01,0.21),
                   fontsize=10, frameon=False)

cax = inset_axes(ax, width="11%", height="2.5%", loc='lower left',
                 bbox_to_anchor=(0.025, 0.18, 1, 1), bbox_transform=ax.transAxes, borderpad=0)

sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cb = plt.colorbar(sm, cax=cax, orientation='horizontal')
cb.set_label('Impervious Area Fraction', fontsize=10)
cb.set_ticks([0.0, 0.5, 1.0])
cb.ax.tick_params(labelsize=8)

plt.show()
```
