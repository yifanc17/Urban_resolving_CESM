### Fig 4a

```python
def extract_r(dict_obj):
    ctrl_r = [v['ctrl_r'] for v in dict_obj.values() if not np.isnan(v['ctrl_r'])]
    test_r = [v['test_r'] for v in dict_obj.values() if not np.isnan(v['test_r'])]
    return ctrl_r, test_r

humid_ctrl, humid_test = extract_r(humid_city_JJA_dict)
temp_ctrl, temp_test = extract_r(temperate_city_JJA_dict)
semi_ctrl, semi_test = extract_r(semiarid_city_JJA_dict)
arid_ctrl, arid_test = extract_r(arid_city_JJA_dict)

data = [humid_ctrl, humid_test,
        temp_ctrl, temp_test,
        semi_ctrl, semi_test,
        arid_ctrl, arid_test]

positions = [1, 2,   4, 5,   7, 8,   10, 11]
colors = ['#9fc5e8', '#f4cccc'] * 4
group_labels = ['Humid', 'Temperate', 'Semi-arid', 'Arid']
label_positions = [1.5, 4.5, 7.5, 10.5]

fig=plt.figure(figsize=(10, 6))
box = plt.boxplot(data, positions=positions, patch_artist=True, widths=0.6,
                  medianprops=dict(color='black', linewidth=1.5), whis=[1, 99], showfliers=False)

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
    
plt.xticks(label_positions, group_labels, fontsize=20)
plt.ylabel('Correlation coefficient', fontsize=24)
plt.yticks(fontsize=20)

legend_patches = [mpatches.Patch(color='#9fc5e8', label='CTRL'),
                  mpatches.Patch(color='#f4cccc', label='USURF')]
plt.legend(handles=legend_patches, loc='upper right', fontsize=18, 
          bbox_to_anchor=(1.015, 1.025), 
            frameon=True,
            edgecolor='grey',
            facecolor='white',
            framealpha=0.6)

plt.tight_layout()
plt.show()
```
