```python
import geocat.viz as gv
from matplotlib.collections import PathCollection
```

```python
rad_vars = ["SWup", "LWup", "Q*", "Qle", "Qh", "Qs"]

sites = [
    "AU-Preston", "AU-SurreyHills", "CA-Sunset", "FI-Kumpula", "FI-Torni", "FR-Capitole",
    "GR-HECKOR", "JP-Yoyogi", "KR-Jungnang", "KR-Ochang", "MX-Escandon", "NL-Amsterdam",
    "PL-Lipowa", "PL-Narutowicza", "SG-TelokKurau", "UK-KingsCollege", "UK-Swindon",
    "US-Baltimore", "US-Minneapolis1", "US-Minneapolis2", "US-WestPhoenix"
]

sites_less = [s for s in sites if s != "MX-Escandon"]
freq_mismatch_sites = ["JP-Yoyogi", "PL-Lipowa", "PL-Narutowicza", "US-Baltimore", "UK-KingsCollege", "UK-Swindon"]

markers = ['o', 's', '^', 'v', '<', '>', '1', '2', '3', '4', 'p', 'P', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']
markers_less = [m for m in markers if m != 'p']
```

```python
def calculate_daily_average(df): # not necessarily includling all hours from a day, just average of all available data
    df['date'] = df['time'].dt.date
    daily_avg = df.groupby('date')[['obs','ctrl', 'detailed', 'test']].mean().reset_index()
    return daily_avg
```

```python
def centered_rmsd(sdev, ccorr):
    return np.sqrt(sdev**2 + 1 - 2*sdev*ccorr)

def calculate_metrics(data_list, obs_list, sdev_ref):
    sdev = np.nanstd(data_list) / sdev_ref
    ccorr = np.corrcoef(data_list, obs_list)[0, 1]
    crmsd = centered_rmsd(sdev, ccorr)
    return sdev, ccorr, crmsd

sites = [
    "AU-Preston", "AU-SurreyHills", "CA-Sunset", "FI-Kumpula", "FI-Torni", "FR-Capitole",
    "GR-HECKOR", "JP-Yoyogi", "KR-Jungnang", "KR-Ochang", "MX-Escandon", "NL-Amsterdam",
    "PL-Lipowa", "PL-Narutowicza", "SG-TelokKurau", "UK-KingsCollege", "UK-Swindon",
    "US-Baltimore", "US-Minneapolis1", "US-Minneapolis2", "US-WestPhoenix"
]

sites_less = [s for s in sites if s != "MX-Escandon"]

def load_and_process_data(var_index, daily=True):
    var = rad_vars[var_index]
    ctrl_sdev_list, test_sdev_list, detailed_sdev_list = [], [], []
    ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list = [], [], []
    ctrl_crmsd_list, test_crmsd_list, detailed_crmsd_list = [], [], []
    use_reduced_sites = var_index in [1, 2, 5]
    site_list = sites_less if use_reduced_sites else sites
    for site in site_list:
        df = pd.read_csv(f'/glade/campaign/univ/uiuc0056/yifan/03_urban_plumber/{site}_{var}_timeseries_wtimestep.csv', parse_dates=['time'])

        if daily:
            df = calculate_daily_average(df)
        
        ctrl_list = df['ctrl'].dropna().tolist()
        test_list = df['test'].dropna().tolist()
        detailed_list = df['detailed'].dropna().tolist()
        obs_list = df['obs'].dropna().tolist()
        
        sdev_ref = np.nanstd(obs_list)
        
        ctrl_sdev, ctrl_ccorr, ctrl_crmsd = calculate_metrics(ctrl_list, obs_list, sdev_ref)
        test_sdev, test_ccorr, test_crmsd = calculate_metrics(test_list, obs_list, sdev_ref)
        detailed_sdev, detailed_ccorr, detailed_crmsd = calculate_metrics(detailed_list, obs_list, sdev_ref)
        
        if all(~np.isnan([ctrl_sdev, test_sdev, detailed_sdev])):
            ctrl_sdev_list.extend([ctrl_sdev])
            test_sdev_list.extend([test_sdev])
            detailed_sdev_list.extend([detailed_sdev])
            
        if all(~np.isnan([ctrl_ccorr, test_ccorr, detailed_ccorr])):
            ctrl_ccorr_list.extend([ctrl_ccorr])
            test_ccorr_list.extend([test_ccorr])
            detailed_ccorr_list.extend([detailed_ccorr])
            
        if all(~np.isnan([ctrl_crmsd, test_crmsd, detailed_crmsd])):
            ctrl_crmsd_list.extend([ctrl_crmsd])
            test_crmsd_list.extend([test_crmsd])
            detailed_crmsd_list.extend([detailed_crmsd])
    
    return (ctrl_sdev_list, test_sdev_list, detailed_sdev_list,
            ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list,
            ctrl_crmsd_list, test_crmsd_list, detailed_crmsd_list)
```

```python
def style_taylor_diagram(taylor):
    for collection in taylor.ax.collections:
        if isinstance(collection, PathCollection):
            collection.set_edgecolor(collection.get_edgecolors())
            collection.set_facecolor('none')
            collection.set_linewidth(2)
            collection.set_sizes([250])
    
    taylor.ax.tick_params(width=2)
    taylor.ax.set_xlim(-0.2, 1.9)
    taylor.ax.set_ylim(-0.2, 1.9)
    
    for text in taylor.ax.texts:
        text.set_visible(False)
    
    taylor.add_contours(
        levels=np.arange(0, 1.1, 0.2),
        colors='lightgray',
        linestyles='--',
        linewidth=1
    )
    
    taylor.set_fontsizes_and_pad(
        ticklabel_fontsize=24,
        axislabel_fontsize=30,
        axislabel_pad=10
    )
```

```python
def create_taylor_plot(site_list, marker_list, sdev_lists, ccorr_lists, test_label='USURF'):
    fig = plt.figure(figsize=(10, 10))
    taylor = gv.TaylorDiagram(fig=fig, label='REF', std_range=(0, 1.8))
    taylor.add_xgrid(np.array([0.6, 0.9]))
    
    ctrl_sdev_list, test_sdev_list, detailed_sdev_list = sdev_lists
    ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list = ccorr_lists
    
    colors = ['#9fc5e8', '#f4cccc', '#ffe599']
    labels = ['CTRL', test_label, 'DETAILED']
    
    for i, site in enumerate(site_list):
        mk = marker_list[i]
        data_sets = [(ctrl_sdev_list[i], ctrl_ccorr_list[i]),
                     (test_sdev_list[i], test_ccorr_list[i]),
                     (detailed_sdev_list[i], detailed_ccorr_list[i])]
        
        for j, (sdev, ccorr) in enumerate(data_sets):
            taylor.add_model_set(
                stddev=[sdev],
                corrcoef=[ccorr],
                label=labels[j],
                color=colors[j],
                marker=mk
            )
    
    style_taylor_diagram(taylor)
    return fig, taylor
```

```python
for var_index in range(6):
    data = load_and_process_data(var_index, daily=True)
    ctrl_sdev_list, test_sdev_list, detailed_sdev_list = data[:3]
    ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list = data[3:6]
    ctrl_crmsd_list, test_crmsd_list, detailed_crmsd_list = data[6:]
    
    use_reduced_sites = var_index in [1, 2, 5]
    site_list = sites_less if use_reduced_sites else sites
    marker_list = markers_less if use_reduced_sites else markers
    
    fig, taylor = create_taylor_plot(
        site_list, marker_list,
        (ctrl_sdev_list, test_sdev_list, detailed_sdev_list),
        (ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list)
    )
```

```python
ctrl_sdev_list=[]
test_sdev_list=[]
detailed_sdev_list=[]
ctrl_ccorr_list=[]
test_ccorr_list=[]
detailed_ccorr_list=[]
ctrl_crmse_list=[]
test_crmse_list=[]
detailed_crmse_list=[]

var=rad_vars[4]
daily=True
use_reduced_sites = var_index in [1, 2, 5]
site_list = sites_less if use_reduced_sites else sites
for site in site_list:
    df = pd.read_csv(f'/glade/campaign/univ/uiuc0056/yifan/03_urban_plumber/{site}_{var}_timeseries_wtimestep.csv', parse_dates=['time'])

    if daily:
        df = calculate_daily_average(df)
    
    ctrl_list = df['ctrl'].dropna().tolist()
    test_list = df['test'].dropna().tolist()
    detailed_list = df['detailed'].dropna().tolist()
    obs_list = df['obs'].dropna().tolist()
    def centered_RMSE(sdev, ccorr):
        return np.sqrt(sdev**2+1-2*sdev*ccorr)
    
    sdev_ref = np.nanstd(obs_list)
    
    ctrl_sdev=np.nanstd(ctrl_list)/sdev_ref
    test_sdev=np.nanstd(test_list)/sdev_ref
    detailed_sdev=np.nanstd(detailed_list)/sdev_ref
    
    ctrl_ccorr=np.corrcoef(ctrl_list, obs_list)[0, 1]
    test_ccorr=np.corrcoef(test_list, obs_list)[0, 1]
    detailed_ccorr=np.corrcoef(detailed_list, obs_list)[0, 1]
    
    ctrl_crmse=centered_RMSE(ctrl_sdev,ctrl_ccorr)
    test_crmse=centered_RMSE(test_sdev,test_ccorr)
    detailed_crmse=centered_RMSE(detailed_sdev,detailed_ccorr)

    if ~np.isnan(ctrl_sdev) and ~np.isnan(test_sdev) and ~np.isnan(detailed_sdev): 
        ctrl_sdev_list.append(ctrl_sdev)
        test_sdev_list.append(test_sdev)
        detailed_sdev_list.append(detailed_sdev)
    if ~np.isnan(ctrl_ccorr) and ~np.isnan(test_ccorr) and ~np.isnan(detailed_ccorr):  
        ctrl_ccorr_list.append(ctrl_ccorr)
        test_ccorr_list.append(test_ccorr)
        detailed_ccorr_list.append(detailed_ccorr)
    if ~np.isnan(ctrl_crmse) and ~np.isnan(test_crmse) and ~np.isnan(detailed_crmse):    
        ctrl_crmse_list.append(ctrl_crmse)
        test_crmse_list.append(test_crmse)
        detailed_crmse_list.append(detailed_crmse)
```

```python
ctrl_sdev_mean = np.mean(ctrl_sdev_list)
test_sdev_mean = np.mean(test_sdev_list)
detailed_sdev_mean = np.mean(detailed_sdev_list)

ctrl_ccorr_mean = np.mean(ctrl_ccorr_list)
test_ccorr_mean = np.mean(test_ccorr_list)
detailed_ccorr_mean = np.mean(detailed_ccorr_list)

ctrl_crmse_mean = np.mean(ctrl_crmse_list)
test_crmse_mean = np.mean(test_crmse_list)
detailed_crmse_mean = np.mean(detailed_crmse_list)

if test_sdev_mean > 1 and ctrl_sdev_mean > 1:
    sdev_improvement = ((test_sdev_mean - ctrl_sdev_mean) / ctrl_sdev_mean) * 100 # negative is better
elif test_sdev_mean < 1 and ctrl_sdev_mean < 1:
    sdev_improvement = ((ctrl_sdev_mean - test_sdev_mean) / ctrl_sdev_mean) * 100 # negative is better
ccorr_improvement = ((test_ccorr_mean - ctrl_ccorr_mean) / ctrl_ccorr_mean) * 100
crmse_improvement = ((ctrl_crmse_mean - test_crmse_mean) / ctrl_crmse_mean) * 100  # Lower is better for RMSE

sdev_improved_sites = sum(1 for i in range(len(test_sdev_list)) if test_sdev_list[i] > ctrl_sdev_list[i])
ccorr_improved_sites = sum(1 for i in range(len(test_ccorr_list)) if test_ccorr_list[i] > ctrl_ccorr_list[i])
crmse_improved_sites = sum(1 for i in range(len(test_crmse_list)) if test_crmse_list[i] < ctrl_crmse_list[i])

sdev_improvement_pct = (sdev_improved_sites / len(test_sdev_list)) * 100
ccorr_improvement_pct = (ccorr_improved_sites / len(test_ccorr_list)) * 100
crmse_improvement_pct = (crmse_improved_sites / len(test_crmse_list)) * 100

print(f"Variable: {var}")
print(f"Number of sites analyzed: {len(test_sdev_list)}")
print("\n=== MEAN VALUES ===")
print(f"Normalized Standard Deviation:")
print(f"  CTRL: {ctrl_sdev_mean:.3f}, USURF: {test_sdev_mean:.3f}, DETAILED: {detailed_sdev_mean:.3f}")
print(f"Correlation Coefficient:")
print(f"  CTRL: {ctrl_ccorr_mean:.3f}, USURF: {test_ccorr_mean:.3f}, DETAILED: {detailed_ccorr_mean:.3f}")
print(f"Centered RMSE:")
print(f"  CTRL: {ctrl_crmse_mean:.3f}, USURF: {test_crmse_mean:.3f}, DETAILED: {detailed_crmse_mean:.3f}")

print(f"\n=== IMPROVEMENTS (USURF vs CTRL) ===")
print(f"Standard Deviation: {sdev_improvement:+.1f}% change")
print(f"Correlation: {ccorr_improvement:+.1f}% improvement")
print(f"Centered RMSE: {crmse_improvement:+.1f}% reduction")

print(f"\n=== SITES SHOWING IMPROVEMENT ===")
print(f"Standard Deviation: {sdev_improved_sites}/{len(test_sdev_list)} sites ({sdev_improvement_pct:.1f}%)")
print(f"Correlation: {ccorr_improved_sites}/{len(test_ccorr_list)} sites ({ccorr_improvement_pct:.1f}%)")
print(f"Centered RMSE: {crmse_improved_sites}/{len(test_crmse_list)} sites ({crmse_improvement_pct:.1f}%)")

sdev_diff = np.array(test_sdev_list) - np.array(ctrl_sdev_list)
ccorr_diff = np.array(test_ccorr_list) - np.array(ctrl_ccorr_list)
crmse_diff = np.array(ctrl_crmse_list) - np.array(test_crmse_list)  

sdev_ci = stats.t.interval(0.95, len(sdev_diff)-1, loc=np.mean(sdev_diff), scale=stats.sem(sdev_diff))
ccorr_ci = stats.t.interval(0.95, len(ccorr_diff)-1, loc=np.mean(ccorr_diff), scale=stats.sem(ccorr_diff))
crmse_ci = stats.t.interval(0.95, len(crmse_diff)-1, loc=np.mean(crmse_diff), scale=stats.sem(crmse_diff))

print(f"\n=== 95% CONFIDENCE INTERVALS ===")
print(f"Standard Deviation change: {sdev_ci[0]:.3f} to {sdev_ci[1]:.3f}")
print(f"Correlation improvement: {ccorr_ci[0]:.3f} to {ccorr_ci[1]:.3f}")
print(f"Centered RMSE reduction: {crmse_ci[0]:.3f} to {crmse_ci[1]:.3f}")
```
