```python
rad_vars = ["SWup", "LWup", "Q*", "Qle", "Qh", "Qs"]

sites = [
    "AU-Preston", "AU-SurreyHills", "CA-Sunset", "FI-Kumpula", "FI-Torni", "FR-Capitole",
    "GR-HECKOR", "JP-Yoyogi", "KR-Jungnang", "KR-Ochang", "MX-Escandon", "NL-Amsterdam",
    "PL-Lipowa", "PL-Narutowicza", "SG-TelokKurau", "UK-KingsCollege", "UK-Swindon",
    "US-Baltimore", "US-Minneapolis1", "US-Minneapolis2", "US-WestPhoenix"
]

sites_less = [s for s in sites if s != "MX-Escandon"]
freq_mismatch_sites = ["JP-Yoyogi", "PL-Lipowa", "PL-Narutowicza", "US-Baltimore", "UK-KingsCollege", "UK-Swindon"]

markers = ['o', 's', '^', 'v', '<', '>', '1', '2', '3', '4', 'p', 'P', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']
markers_less = [m for m in markers if m != 'p']
```

```python
def calc_rolling_7d(site, type_name):
    series = pd.Series(eval(f'{type_name}_list'))
    window = 168 if site in freq_mismatch_sites else 336
    return series.rolling(window=window, center=True).mean().dropna().tolist()

def centered_rmsd(sdev, ccorr):
    return np.sqrt(sdev**2 + 1 - 2*sdev*ccorr)

def calculate_metrics(data_list, obs_list, sdev_ref):
    sdev = np.nanstd(data_list) / sdev_ref
    ccorr = np.corrcoef(data_list, obs_list)[0, 1]
    crmsd = centered_rmsd(sdev, ccorr)
    return sdev, ccorr, crmsd

def load_and_process_data(var_index, rolling=False):
    var = rad_vars[var_index]
    ctrl_sdev_list, test_sdev_list, detailed_sdev_list = [], [], []
    ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list = [], [], []
    ctrl_crmsd_list, test_crmsd_list, detailed_crmsd_list = [], [], []
    
    for site in sites:
        df = pd.read_csv(f'/glade/campaign/univ/uiuc0056/yifan/03_urban_plumber/{site}_{var}_timeseries.csv')
        
        ctrl_list = df['ctrl'].dropna().tolist()
        test_list = df['test'].dropna().tolist()
        detailed_list = df['detailed'].dropna().tolist()
        obs_list = df['obs'].dropna().tolist()

        if rolling:
            ctrl_list = calc_rolling_7d(ctrl, site)
            test_list = calc_rolling_7d(test, site)
            detailed_list = calc_rolling_7d(detailed, site)
            obs_list = calc_rolling_7d(obs, site)
        
        sdev_ref = np.nanstd(obs_list)
        
        ctrl_sdev, ctrl_ccorr, ctrl_crmsd = calculate_metrics(ctrl_list, obs_list, sdev_ref)
        test_sdev, test_ccorr, test_crmsd = calculate_metrics(test_list, obs_list, sdev_ref)
        detailed_sdev, detailed_ccorr, detailed_crmsd = calculate_metrics(detailed_list, obs_list, sdev_ref)
        
        if all(~np.isnan([ctrl_sdev, test_sdev, detailed_sdev])):
            ctrl_sdev_list.extend([ctrl_sdev])
            test_sdev_list.extend([test_sdev])
            detailed_sdev_list.extend([detailed_sdev])
            
        if all(~np.isnan([ctrl_ccorr, test_ccorr, detailed_ccorr])):
            ctrl_ccorr_list.extend([ctrl_ccorr])
            test_ccorr_list.extend([test_ccorr])
            detailed_ccorr_list.extend([detailed_ccorr])
            
        if all(~np.isnan([ctrl_crmsd, test_crmsd, detailed_crmsd])):
            ctrl_crmsd_list.extend([ctrl_crmsd])
            test_crmsd_list.extend([test_crmsd])
            detailed_crmsd_list.extend([detailed_crmsd])
    
    return (ctrl_sdev_list, test_sdev_list, detailed_sdev_list,
            ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list,
            ctrl_crmsd_list, test_crmsd_list, detailed_crmsd_list)
```

```python
def style_taylor_diagram(taylor):
    for collection in taylor.ax.collections:
        if isinstance(collection, PathCollection):
            collection.set_edgecolor(collection.get_edgecolors())
            collection.set_facecolor('none')
            collection.set_linewidth(2)
            collection.set_sizes([250])
    
    taylor.ax.tick_params(width=2)
    taylor.ax.set_xlim(-0.2, 1.9)
    taylor.ax.set_ylim(-0.2, 1.9)
    
    for text in taylor.ax.texts:
        text.set_visible(False)
    
    taylor.add_contours(
        levels=np.arange(0, 1.1, 0.2),
        colors='lightgray',
        linestyles='--',
        linewidth=1
    )
    
    taylor.set_fontsizes_and_pad(
        ticklabel_fontsize=24,
        axislabel_fontsize=30,
        axislabel_pad=10
    )
```

```python
def create_taylor_plot(site_list, marker_list, sdev_lists, ccorr_lists, test_label='USURF'):
    fig = plt.figure(figsize=(10, 10))
    taylor = gv.TaylorDiagram(fig=fig, label='REF', std_range=(0, 1.8))
    taylor.add_xgrid(np.array([0.6, 0.9]))
    
    ctrl_sdev_list, test_sdev_list, detailed_sdev_list = sdev_lists
    ctrl_ccorr_list, test_ccorr_list, detailed_ccorr_list = ccorr_lists
    
    colors = ['#9fc5e8', '#f4cccc', '#ffe599']
    labels = ['CTRL', test_label, 'DETAILED']
    
    for i, site in enumerate(site_list):
        mk = marker_list[i]
        data_sets = [(ctrl_sdev_list[i], ctrl_ccorr_list[i]),
                     (test_sdev_list[i], test_ccorr_list[i]),
                     (detailed_sdev_list[i], detailed_ccorr_list[i])]
        
        for j, (sdev, ccorr) in enumerate(data_sets):
            taylor.add_model_set(
                stddev=[sdev],
                corrcoef=[ccorr],
                label=labels[j],
                color=colors[j],
                marker=mk
            )
    
    style_taylor_diagram(taylor)
    return fig, taylor
```
